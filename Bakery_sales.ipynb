## Environment & Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from utilsforecast.plotting import plot_series
from utilsforecast.evaluation import evaluate
from utilsforecast.losses import *

import warnings
warnings.filterwarnings("ignore")
```

## Data Loading

```python
df = pd.read_csv("daily_sales_french_bakery.csv", parse_dates=["ds"])
df = df.groupby('unique_id').filter(lambda x: len(x) >= 28)  # keep series with enough history
df = df.sort_values(["unique_id", "ds"]).reset_index(drop=True)

print(f"Rows: {len(df):,}  |  Unique products: {df['unique_id'].nunique()}")
print("Date range:", df["ds"].min().date(), "→", df["ds"].max().date())
display(df.head())

# Keep a model input copy without exogenous price
df_model = df.drop(columns=["unit_price"], errors="ignore")
                                                                 
```

## Exploratory Analysis
Quick look at the time series to spot trend, seasonality, or anomalies.

```python
TARGET_PRODUCTS = ["BAGUETTE", "CROISSANT", "PAIN AU CHOCOLAT", "MACARON"]
DISPLAY_PRODUCTS = TARGET_PRODUCTS[:2]  

HORIZON = 7  
FREQ = "D"   
```

```python
plot_series(df=df, ids=DISPLAY_PRODUCTS, palette="viridis")
```

```python
plot_series(df=df, ids=DISPLAY_PRODUCTS, max_insample_length=56, palette="viridis")
```

## Baseline Models
Quick, explainable baselines (Naive / Historic / Window / SeasonalNaive) provide a reality check before complex models.


```python
from statsforecast import StatsForecast
from statsforecast.models import Naive, HistoricAverage, WindowAverage,SeasonalNaive
```

```python
horizon = 7

models = [
    Naive(),
    HistoricAverage(),
    WindowAverage(window_size=7),
    SeasonalNaive(season_length = 7)
]

sf = StatsForecast(models=models, freq = "D")
sf.fit(df=df_model)
preds = sf.predict(h=horizon)
```

```python
preds.head()
```

```python
plot_series(
    df=df,
    forecasts_df=preds,
    ids=["BAGUETTE", "CROISSANT"],
    max_insample_length=28,
    palette="viridis")
```

### Evaluate Baselines on a Proper Validation Split
We hold out the last **7 days** per product to avoid leakage, fit on the rest, and compare MAE.


```python
test  = df_model.groupby("unique_id").tail(horizon)
train = df_model.drop(test.index).reset_index(drop=True)
```

```python
sf.fit(df=train)
preds = sf.predict(h=horizon)
eval_df = pd.merge(test, preds, 'left', ['ds', 'unique_id'])
```

```python
evaluation = evaluate(
    eval_df,
    metrics = [mae],
)
evaluation.head()
```

```python
evaluation = evaluation.drop(["unique_id"], axis=1).groupby("metric").mean().reset_index()
evaluation
```

```python
methods = evaluation.columns[1:].tolist()
values = evaluation.iloc[0, 1:]. tolist()

plt. figure(figsize=(10, 6))
bars = plt.bar (methods, values)

for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
             f'{value:.3f}', ha='center', va='bottom', fontweight='bold')
             
plt.xlabel ('Methods')
plt.ylabel('Mean absolute error (MAE)')
plt.tight_layout()
plt.show()
```

## AutoARIMA

```python
from statsforecast.models import AutoARIMA
```

```python
unique_ids = DISPLAY_PRODUCTS 

small_train = train[train["unique_id"].isin(unique_ids)]
small_test = test[test["unique_id"].isin(unique_ids)]

models = [
    AutoARIMA(seasonal=False, alias="ARIMA"),
    AutoARIMA(season_length=7, alias="SARIMA")
]

sf = StatsForecast(models=models, freq="D")
sf.fit(df=small_train)
arima_preds = sf.predict(h=horizon)

arima_eval_df = pd.merge(arima_preds, eval_df, 'inner', ['ds', 'unique_id'])
arima_eval = evaluate(
    arima_eval_df,
    metrics=[mae],
)
arima_eval
```

```python
arima_eval = arima_eval.drop(['unique_id'], axis=1).groupby('metric').mean().reset_index()
arima_eval
```

```python
plot_series(
    df=df_model,
    forecasts_df=arima_preds,
    ids=DISPLAY_PRODUCTS,      
    max_insample_length=28,
    palette="viridis"
)
```

```python
methods = arima_eval.columns[1:].tolist()  
values = arima_eval.iloc[0, 1:].tolist() 

sorted_data = sorted(zip(methods, values), key=lambda x: x[1], reverse=True)
methods_sorted, values_sorted = zip(*sorted_data)

plt.figure(figsize=(10, 6))
bars = plt.bar(methods_sorted, values_sorted)

for bar, value in zip(bars, values_sorted):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05, 
             f'{value:.3f}', ha='center', va='bottom', fontweight='bold')

plt.xlabel('Methods')
plt.ylabel('Mean absolute error (MAE)')
plt.tight_layout()

plt.show()
```

## Cross-validation

```python
small_df = df[df["unique_id"].isin(unique_ids)]

models = [
    SeasonalNaive(season_length=7),
    AutoARIMA(seasonal=False, alias="ARIMA"),
    AutoARIMA(season_length=7, alias="SARIMA")
]

sf = StatsForecast(models=models, freq="D")
cv_df = sf.cross_validation(
    h=horizon, 
    df=small_df,
    n_windows=8,
    step_size=horizon,
    refit=True
)

cv_df.head()
```

```python
plot_series(
    df=small_df, 
    forecasts_df=cv_df.drop(["y", "cutoff"], axis=1), 
    ids=["BAGUETTE", "CROISSANT"], 
    max_insample_length=140, 
    palette="viridis")
```

```python
cv_eval = evaluate(
    cv_df.drop(["cutoff"], axis=1),
    metrics=[mae],
)
cv_eval = cv_eval.drop(['unique_id'], axis=1).groupby('metric').mean().reset_index()
cv_eval
```

```python
methods = cv_eval.columns[1:].tolist()  
values = cv_eval.iloc[0, 1:].tolist() 

sorted_data = sorted(zip(methods, values), key=lambda x: x[1], reverse=True)
methods_sorted, values_sorted = zip(*sorted_data)

plt.figure(figsize=(10, 6))
bars = plt.bar(methods_sorted, values_sorted)

for bar, value in zip(bars, values_sorted):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05, 
             f'{value:.3f}', ha='center', va='bottom', fontweight='bold')

plt.xlabel('Methods')
plt.ylabel('Mean absolute error (MAE)')
plt.tight_layout()

plt.show()
```

## Evaluation metrics

```python
models = [
    AutoARIMA(season_length=7, alias="SARIMA_exog"),
    SeasonalNaive(season_length=7)
]

sf = StatsForecast(models=models, freq="D")
final_cv_df = sf.cross_validation(
    h=horizon,
    df=small_df,
    n_windows=8,
    step_size=7,
    refit=True,
    level=[80],
)
```

```python
final_cv_df.head()
```

```python
temp_test = small_df.groupby("unique_id").tail(7*8)
temp_train = small_df.drop(temp_test.index).reset_index(drop=True)
```

```python
models = ["SARIMA_exog", "SeasonalNaive"]
metrics = [
    mae,
    mse, 
    rmse, 
    mape, 
    smape,
    partial(mase, seasonality=7),
    scaled_crps
]

final_eval = evaluate(
    final_cv_df.drop(["ds", "cutoff"], axis=1),
    metrics=metrics,
    models=models,
    train_df=temp_train,
    level=[80]
)
final_eval = final_eval.drop(['unique_id'], axis=1).groupby('metric').mean().reset_index()
final_eval
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

axes_flat = axes.flatten()

models = ['SARIMA_exog', 'SeasonalNaive']
x_pos = [0, 1]
colors = ['blue', 'red']

for i, row in final_eval.iterrows():
    ax = axes_flat[i]
    
    model_values = [row['SARIMA_exog'], row['SeasonalNaive']]
    
    bars = ax.bar(x_pos, model_values, color=colors, alpha=0.8, edgecolor='black', linewidth=1)
    
    for j, (bar, value) in enumerate(zip(bars, model_values)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{value:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=10)
    
    ax.set_title(row['metric'].upper(), fontweight='bold', fontsize=12)
    ax.set_xticks(x_pos)
    ax.set_xticklabels(models, ha='center')
    ax.set_ylabel('Value')
    max_value = max(model_values)
    ax.set_ylim(0, max_value * 1.1)

fig.delaxes(axes_flat[7])

axes_flat[8].set_visible(False)

plt.tight_layout()
plt.show()
```

## Business Conclusion

This project applied statistical forecasting (baselines and AutoARIMA/SARIMA) to the bakery’s daily sales.

### Key Insights
- **Baseline comparison:** Even simple models (Naïve, Seasonal Naïve) capture part of the weekly pattern, but more advanced approaches reduce forecast errors further.
- **AutoARIMA vs SARIMA:** Seasonal ARIMA consistently outperformed plain ARIMA, confirming the presence of strong **weekly seasonality** in bakery demand.
- **Product demand patterns:**
  - **Croissant & Baguette:** Clear weekend spikes, suggesting higher demand Friday–Sunday.
  - **Other items (e.g., pain au chocolat):** Smaller but similar seasonal lift, with weekday troughs.

### Business Implications
- **Inventory planning:** Increase stock of croissants and baguettes by ~15–20% on weekends to avoid stock-outs.  
- **Staffing:** Schedule additional staff during weekend mornings to handle peak bakery traffic.  
- **Waste reduction:** Use forecasts to set conservative weekday production targets, minimizing over-production and waste.  

### Next Steps
- **External factors:** Incorporate holidays, promotions, and weather as exogenous regressors for more accurate forecasts.  
- **Automation:** Deploy the forecasting pipeline to run weekly, updating inventory recommendations dynamically.  
- **Scaling:** Extend the model loop to all bakery products, producing a consolidated dashboard for managers.  

**Bottom line:**  
By moving from intuition to data-driven forecasting, the bakery can **improve weekend readiness, reduce weekday waste, and optimize staffing**, directly improving both profitability and customer satisfaction.
